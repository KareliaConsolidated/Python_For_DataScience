### Widget Example

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset('mpg')
```

```python
def bandwidth_widget(bw=1):
    sns.kdeplot(cars.horsepower, lw=3, fill=True, bw_adjust=bw)
    plt.xlim(-30, 300)
    plt.ylim(0, 0.03);
```

```python
widgets.interact(bandwidth_widget, bw=(.1, 3));
```

```python
@widgets.interact(bw=(.1, 3, .1))
def bandwidth_widget(bw=1):
    sns.kdeplot(cars.horsepower, lw=3, fill=True, bw_adjust=bw)
    plt.xlim(-30, 300)
    plt.ylim(0, 0.03);
```

```python
import numpy as np
```

```python
x = np.random.uniform(0, 5, size=100)
ep = np.random.normal(size=100)

y = 2*x + ep
```

```python
plt.scatter(x, y);
```

```python
x_values = np.linspace(0, 5, 1000)
```

### Slider

```python
def slope_viz(m=1, line=False, text=""):
    plt.scatter(x, y)
    if line:
        plt.plot(x_values, m*x_values, lw=3, color='black');
    if text:
        plt.text(0, 10, f"{text}\n The Slope is {m}.")
    plt.ylim(-1.2, 12.2)
```

```python
slope_viz(m=4)
```

```python
widgets.interact(slope_viz, m=(0.2, 5, 0.2), line=False, text="");
```

### Matplotlib: 3D Plots

```python
np.random.seed(31)
```

```python
mu = 3
n = 50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
```

```python
plt.rcParams['figure.figsize'] = (8,6)
```

### 3D Scatter Plot

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, s=100);

ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');

ax.view_init(45, 215);
```

### 3D Line Plot

```python
omega = 4
z_line = np.linspace(0, 10, 100)
x_line = np.cos(omega*z_line)
y_line = np.sin(omega*z_line)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x_line, y_line, z_line); 
```
