#### Time Series
A time series is a measure of unit change over time for any variable under observation.

#### Segment - Visualizing Time Series

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

#### The simplest time series plot

```python
address = 'Datasets/Superstore-Sales.csv'

df = pd.read_csv(address, index_col='Order Date', encoding='cp1252', parse_dates = True)
df.head()
```

```python
df['Order Quantity'].plot()
plt.show()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
plt.show()
```
