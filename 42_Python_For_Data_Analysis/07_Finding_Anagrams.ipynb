```python
def signature(word):
    return ''.join(sorted(word))
```

```python
def find_anagram(myword):
    mysig = signature(myword)
    
    for word in words:
        if mysig == signature(word):
            print(word)
```

```python
find_anagram('dictionary')
```

```python
words_by_sig = collections.defaultdict(set)

for word in words:
    words_by_sig[signature(word)].add(word)
```

```python
words_by_sig
```

```python
anagram_by_sig = {sig:wordset for sig,wordset in words_by_sig.items() if len(wordset) > 1}
```

```python
anagram_by_sig
```

```python
def find_anagram_fast(myword):
    sig = signature(myword)
    return anagram_by_sig[sig]
```

```python
find_anagram_fast('tops')
```
