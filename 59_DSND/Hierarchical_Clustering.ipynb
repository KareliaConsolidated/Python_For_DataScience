```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris.data[:10]
```

```python
iris.target[:10]
```

### Clustering

```python
from sklearn.cluster import AgglomerativeClustering

ward = AgglomerativeClustering(n_clusters=3)
ward_pred = ward.fit_predict(iris.data)
```

```python
complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_pred = complete.fit_predict(iris.data)
```

```python
avg = AgglomerativeClustering(n_clusters=3, linkage='average')
avg_pred = avg.fit_predict(iris.data)
```

```python
from sklearn.metrics import adjusted_rand_score

ward_score = adjusted_rand_score(iris.target, ward_pred)

complete_score = adjusted_rand_score(iris.target, complete_pred)

avg_score = adjusted_rand_score(iris.target, avg_pred)

print( "Scores: \nWard:", ward_score,"\nComplete: ", complete_score, "\nAverage: ", avg_score)
```

```python
from sklearn.preprocessing import normalize

normalized_data = normalize(iris.data)

normalized_data[:15]
```

```python
ward_pred = ward.fit_predict(normalized_data)
complete_pred = complete.fit_predict(normalized_data)
avg_pred = avg.fit_predict(normalized_data)

ward_score = adjusted_rand_score(iris.target, ward_pred)
complete_score = adjusted_rand_score(iris.target, complete_pred)
avg_score = adjusted_rand_score(iris.target, avg_pred)

print( "Scores: \nWard:", ward_score,"\nComplete: ", complete_score, "\nAverage: ", avg_score)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

linkage_type = 'ward'

linkage_matrix = linkage(normalized_data, linkage_type)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (22,18))

dendrogram(linkage_matrix)

plt.show()
```
